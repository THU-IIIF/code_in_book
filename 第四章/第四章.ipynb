```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.inspection import permutation_importance
import warnings
warnings.filterwarnings('ignore')

class FuturesDecisionTree:
    """纯碱期货决策树回归模型"""
    
    def __init__(self, data, target_col='收盘价(元)', test_size=0.2, random_state=42):
        self.data = data.copy()
        self.target_col = target_col
        self.test_size = test_size
        self.random_state = random_state
        self.scaler = StandardScaler()
        self.model = None
        self.feature_importance = None
        
    def create_features(self, lookback_days=10):
        """创建技术指标特征"""
        print("创建特征工程...")
        
        df = self.data.copy()
        
        # 确保日期列正确处理
        if '日期' in df.columns:
            df['日期'] = pd.to_datetime(df['日期'])
            df = df.sort_values('日期').reset_index(drop=True)
        
        # 价格特征
        df['高低价比'] = df['最高价(元)'] / df['最低价(元)']
        df['开盘收盘价比'] = df['开盘价(元)'] / df['收盘价(元)']
        
        # 移动平均线特征
        for window in [5, 10, 20]:
            df[f'{window}日移动平均'] = df['收盘价(元)'].rolling(window=window).mean()
            df[f'价格移动平均比_{window}'] = df['收盘价(元)'] / df[f'{window}日移动平均']
            df[f'价格标准差_{window}'] = df['收盘价(元)'].rolling(window=window).std()
        
        # 动量指标
        df['5日动量'] = df['收盘价(元)'] - df['收盘价(元)'].shift(5)
        df['10日动量'] = df['收盘价(元)'] - df['收盘价(元)'].shift(10)
        df['5日价格变化率'] = (df['收盘价(元)'] - df['收盘价(元)'].shift(5)) / df['收盘价(元)'].shift(5) * 100
        df['10日价格变化率'] = (df['收盘价(元)'] - df['收盘价(元)'].shift(10)) / df['收盘价(元)'].shift(10) * 100
        
        # 波动率特征
        df['5日波动率'] = df['收盘价(元)'].rolling(window=5).std() / df['收盘价(元)'].rolling(window=5).mean()
        df['10日波动率'] = df['收盘价(元)'].rolling(window=10).std() / df['收盘价(元)'].rolling(window=10).mean()
        
        # 价格位置特征
        df['20日价格位置'] = (df['收盘价(元)'] - df['收盘价(元)'].rolling(20).min()) / \
                           (df['收盘价(元)'].rolling(20).max() - df['收盘价(元)'].rolling(20).min())
        
        # 涨跌幅特征
        if '涨跌幅' in df.columns:
            df['涨跌幅_绝对值'] = np.abs(df['涨跌幅'])
            df['涨跌幅_5日平均'] = df['涨跌幅'].rolling(5).mean()
            df['涨跌幅_10日平均'] = df['涨跌幅'].rolling(10).mean()
        
        # 滞后特征
        for lag in range(1, lookback_days + 1):
            df[f'收盘价滞后{lag}天'] = df['收盘价(元)'].shift(lag)
        
        # 成交量特征
        if '成交量' in df.columns:
            df['成交量_5日平均'] = df['成交量'].rolling(5).mean()
            df['成交量比率'] = df['成交量'] / df['成交量_5日平均']
            # 价格成交量相关性
            df['价量相关性_5日'] = df['收盘价(元)'].rolling(5).corr(df['成交量'])
        
        # 成交额特征
        if '成交额(百万)' in df.columns:
            df['成交额_5日平均'] = df['成交额(百万)'].rolling(5).mean()
            df['成交额比率'] = df['成交额(百万)'] / df['成交额_5日平均']
            df['平均成交价格'] = df['成交额(百万)'] * 1000000 / df['成交量']  # 计算平均成交价格
        
        # 目标变量：未来1天的收盘价
        df['目标收盘价'] = df['收盘价(元)'].shift(-1)
        
        # 删除含有NaN的行
        df = df.dropna()
        
        print(f"特征创建完成，最终数据形状: {df.shape}")
        return df
    
    def prepare_data(self, lookback_days=10):
        """准备训练数据"""
        print("准备训练数据...")
        
        # 创建特征
        df = self.create_features(lookback_days)
        
        # 分离特征和目标变量
        exclude_cols = ['目标收盘价', '日期', self.target_col]
        feature_cols = [col for col in df.columns if col not in exclude_cols]
        self.feature_cols = feature_cols
        
        X = df[feature_cols]
        y = df['目标收盘价']
        
        print(f"特征数量: {len(feature_cols)}")
        print("前10个特征:", feature_cols[:10])
        
        # 数据分割（时间序列分割，不随机打乱）
        split_idx = int(len(X) * (1 - self.test_size))
        
        self.X_train = X.iloc[:split_idx]
        self.X_test = X.iloc[split_idx:]
        self.y_train = y.iloc[:split_idx]
        self.y_test = y.iloc[split_idx:]
        
        # # 特征标准化
        # self.X_train_scaled = self.scaler.fit_transform(self.X_train)
        # self.X_test_scaled = self.scaler.transform(self.X_test)
        self.X_train_scaled = self.X_train
        self.X_test_scaled = self.X_test
        print(f"训练集大小: {self.X_train_scaled.shape}")
        print(f"测试集大小: {self.X_test_scaled.shape}")
        print(f"训练集时间范围: {df['日期'].iloc[0]} 到 {df['日期'].iloc[split_idx-1]}")
        print(f"测试集时间范围: {df['日期'].iloc[split_idx]} 到 {df['日期'].iloc[-1]}")
        
        return self.X_train_scaled, self.X_test_scaled, self.y_train, self.y_test
    
    def tune_hyperparameters(self):
        """超参数调优"""
        print("进行超参数调优...")
        
        param_grid = {
            'max_depth': [3, 5, 7, 10, 15, None],
            'min_samples_split': [2, 5, 10, 20],
            'min_samples_leaf': [1, 2, 5, 10],
            'max_features': ['auto', 'sqrt', 'log2', None]
        }
        
        dt = DecisionTreeRegressor(random_state=self.random_state)
        grid_search = GridSearchCV(
            dt, param_grid, cv=5, scoring='neg_mean_squared_error', 
            n_jobs=-1, verbose=1
        )
        
        grid_search.fit(self.X_train_scaled, self.y_train)
        
        print(f"最佳参数: {grid_search.best_params_}")
        print(f"最佳交叉验证分数: {-grid_search.best_score_:.4f}")
        
        self.best_params = grid_search.best_params_
        return grid_search.best_params_
    
    def train_model(self, use_grid_search=True, **kwargs):
        """训练决策树模型"""
        print("训练决策树回归模型...")
        
        if use_grid_search and hasattr(self, 'best_params'):
            model_params = self.best_params
        else:
            # 默认参数
            model_params = {
                'max_depth': 10,
                'min_samples_split': 10,
                'min_samples_leaf': 5,
                'max_features': 'sqrt',
                'random_state': self.random_state
            }
            model_params.update(kwargs)
        
        self.model = DecisionTreeRegressor(**model_params)
        self.model.fit(self.X_train_scaled, self.y_train)
        
        # 预测
        self.y_train_pred = self.model.predict(self.X_train_scaled)
        self.y_test_pred = self.model.predict(self.X_test_scaled)
        
        # 计算特征重要性
        self.feature_importance = pd.DataFrame({
            'feature': self.feature_cols,
            'importance': self.model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        return self.model
    
    def evaluate_model(self):
        """评估模型性能"""
        print("\n模型性能评估:")
        print("=" * 50)
        
        # 计算各项指标
        train_mse = mean_squared_error(self.y_train, self.y_train_pred)
        test_mse = mean_squared_error(self.y_test, self.y_test_pred)
        train_mae = mean_absolute_error(self.y_train, self.y_train_pred)
        test_mae = mean_absolute_error(self.y_test, self.y_test_pred)
        train_r2 = r2_score(self.y_train, self.y_train_pred)
        test_r2 = r2_score(self.y_test, self.y_test_pred)
        
        # 计算方向准确性（涨跌方向预测）
        train_direction = np.mean((np.diff(self.y_train.values) * np.diff(self.y_train_pred)) > 0)
        test_direction = np.mean((np.diff(self.y_test.values) * np.diff(self.y_test_pred)) > 0)
        
        # 计算平均预测误差百分比
        train_mape = np.mean(np.abs((self.y_train - self.y_train_pred) / self.y_train)) * 100
        test_mape = np.mean(np.abs((self.y_test - self.y_test_pred) / self.y_test)) * 100
        
        print(f"训练集 - MSE: {train_mse:.4f}, MAE: {train_mae:.4f}, R²: {train_r2:.4f}")
        print(f"训练集 - MAPE: {train_mape:.2f}%, 方向准确率: {train_direction:.4f}")
        print(f"测试集 - MSE: {test_mse:.4f}, MAE: {test_mae:.4f}, R²: {test_r2:.4f}")
        print(f"测试集 - MAPE: {test_mape:.2f}%, 方向准确率: {test_direction:.4f}")
        
        return {
            'train_mse': train_mse,
            'test_mse': test_mse,
            'train_mae': train_mae,
            'test_mae': test_mae,
            'train_r2': train_r2,
            'test_r2': test_r2,
            'train_mape': train_mape,
            'test_mape': test_mape,
            'train_direction': train_direction,
            'test_direction': test_direction
        }
    
    def plot_predictions(self):
        """绘制预测结果"""
        plt.figure(figsize=(15, 12))
        
        # 训练集预测结果
        plt.subplot(3, 1, 1)
        plt.plot(self.y_train.values, label='实际收盘价', alpha=0.8, linewidth=1)
        plt.plot(self.y_train_pred, label='预测收盘价', alpha=0.8, linewidth=1)
        plt.title('训练集: 实际值 vs 预测值')
        plt.xlabel('样本索引')
        plt.ylabel('价格(元)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # 测试集预测结果
        plt.subplot(3, 1, 2)
        plt.plot(self.y_test.values, label='实际收盘价', alpha=0.8, linewidth=1)
        plt.plot(self.y_test_pred, label='预测收盘价', alpha=0.8, linewidth=1)
        plt.title('测试集: 实际值 vs 预测值')
        plt.xlabel('样本索引')
        plt.ylabel('价格(元)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # 散点图
        plt.subplot(3, 2, 5)
        plt.scatter(self.y_test, self.y_test_pred, alpha=0.6, s=20)
        plt.plot([self.y_test.min(), self.y_test.max()], 
                [self.y_test.min(), self.y_test.max()], 'r--', lw=2)
        plt.xlabel('实际收盘价(元)')
        plt.ylabel('预测收盘价(元)')
        plt.title('测试集: 实际值 vs 预测值散点图')
        plt.grid(True, alpha=0.3)
        
        # 残差图
        plt.subplot(3, 2, 6)
        residuals = self.y_test - self.y_test_pred
        plt.scatter(self.y_test_pred, residuals, alpha=0.6, s=20)
        plt.axhline(y=0, color='r', linestyle='--')
        plt.xlabel('预测收盘价(元)')
        plt.ylabel('残差')
        plt.title('测试集残差图')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    # def plot_feature_importance(self, top_n=15):
    #     """绘制特征重要性图"""
    #     if self.feature_importance is None:
    #         print("请先训练模型!")
    #         return
        
    #     plt.figure(figsize=(12, 10))
        
    #     top_features = self.feature_importance.head(top_n)
        
    #     plt.subplot(2, 1, 1)
    #     plt.barh(range(len(top_features)), top_features['importance'], color='skyblue')
    #     plt.yticks(range(len(top_features)), top_features['feature'])
    #     plt.xlabel('特征重要性')
    #     plt.title(f'纯碱期货决策树特征重要性 (Top {top_n})')
    #     plt.gca().invert_yaxis()
    #     plt.grid(True, alpha=0.3, axis='x')
        
    #     plt.subplot(2, 1, 2)
    #     # 累积重要性
    #     cumulative_importance = top_features['importance'].cumsum()
    #     plt.plot(range(1, len(top_features) + 1), cumulative_importance, 'o-', color='orange')
    #     plt.axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='80% 重要性')
    #     plt.axhline(y=0.9, color='green', linestyle='--', alpha=0.7, label='90% 重要性')
    #     plt.xlabel('特征数量')
    #     plt.ylabel('累积重要性')
    #     plt.title('特征累积重要性')
    #     plt.legend()
    #     plt.grid(True, alpha=0.3)
        
    #     plt.tight_layout()
    #     plt.show()
        
    #     print(f"\n前{top_n}个重要特征:")
    #     print(top_features.round(4))
        
    #     # 计算达到80%和90%重要性所需的特征数量
    #     cumulative_all = self.feature_importance['importance'].cumsum()
    #     n_80 = len(cumulative_all[cumulative_all <= 0.8]) + 1
    #     n_90 = len(cumulative_all[cumulative_all <= 0.9]) + 1
    #     print(f"达到80%重要性所需特征数: {n_80}")
    #     print(f"达到90%重要性所需特征数: {n_90}")
    
    def plot_feature_importance(self, top_n=15):
        """绘制特征重要性图"""
        if self.feature_importance is None:
            print("请先训练模型!")
            return
        
        plt.figure(figsize=(12, 10))
        
        top_features = self.feature_importance.head(top_n)
        
        plt.subplot(2, 1, 1)
        plt.barh(range(len(top_features)), top_features['importance'], color='skyblue')
        plt.yticks(range(len(top_features)), top_features['feature'], fontsize=16, fontweight='bold')
        plt.xlabel('特征重要性', fontsize=26, fontweight='bold')
        plt.title(f'纯碱期货决策树特征重要性 (Top {top_n})', fontsize=26, fontweight='bold')
        plt.legend(fontsize=16, loc='best')
        plt.gca().invert_yaxis()
        plt.grid(True, alpha=0.3, axis='x')
        
        plt.subplot(2, 1, 2)
        # 累积重要性
        cumulative_importance = top_features['importance'].cumsum()
        plt.plot(range(1, len(top_features) + 1), cumulative_importance, 'o-', color='orange')
        plt.axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='80% 重要性')
        plt.axhline(y=0.9, color='green', linestyle='--', alpha=0.7, label='90% 重要性')
        plt.xlabel('特征数量', fontsize=26, fontweight='bold')
        plt.ylabel('累积重要性', fontsize=26, fontweight='bold')
        plt.title('特征累积重要性', fontsize=26, fontweight='bold')

    # ax1.legend()
        plt.legend(fontsize=16, loc='best')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('./4-5.png')
        plt.show()
        
        print(f"\n前{top_n}个重要特征:")
        print(top_features.round(4))
        
        # 计算达到80%和90%重要性所需的特征数量
        cumulative_all = self.feature_importance['importance'].cumsum()
        n_80 = len(cumulative_all[cumulative_all <= 0.8]) + 1
        n_90 = len(cumulative_all[cumulative_all <= 0.9]) + 1
        print(f"达到80%重要性所需特征数: {n_80}")
        print(f"达到90%重要性所需特征数: {n_90}")
    def plot_decision_tree(self, max_depth=3):
        """绘制决策树结构（限制深度以便可视化）"""
        if self.model is None:
            print("请先训练模型!")
            return
        
        plt.figure(figsize=(20, 10))
        plot_tree(self.model, 
                 feature_names=self.feature_cols,
                 filled=True, 
                 rounded=True,
                 max_depth=max_depth,
                 fontsize=26,
                 proportion=True)
        # plt.title(f'纯碱期货决策树结构 (最大深度: {max_depth})')
        # plt.title(f'纯碱期货价格预测决策树 (深度限制: {max_depth})', 
        #                 fontsize=26, fontweight='bold', pad=30)
        plt.tight_layout()
        plt.savefig('./4-1.png')
        plt.show()


    def plot_feature_space_stratification(self, feature1='开盘价(元)', feature2='成交量', 
                                        target_col='收盘价(元)', figsize=(16, 12)):
        """绘制基于特征空间分层的预测图（开盘价 vs 成交量）"""
        if self.model is None:
            print("请先训练模型!")
            return
        
        # 准备数据
        df = self.data.copy()
        
        # 选择两个特征和目标变量
        X_plot = df[[feature1, feature2]].dropna()
        y_plot = df[target_col].loc[X_plot.index]
        
        # 训练一个专门用于可视化的决策树
        dt_visual = DecisionTreeRegressor(max_depth=4, random_state=42)
        dt_visual.fit(X_plot, y_plot)
        
        # 创建网格数据
        x_min, x_max = X_plot[feature1].min() - 1, X_plot[feature1].max() + 1
        y_min, y_max = X_plot[feature2].min() - 1, X_plot[feature2].max() + 1
        
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                            np.linspace(y_min, y_max, 200))
        
        # 预测整个网格
        Z = dt_visual.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        # 创建图形
        fig, axes = plt.subplots(2, 2, figsize=figsize)
        
        # 图1: 决策树分区边界
        ax1 = axes[0, 0]
        contour = ax1.contourf(xx, yy, Z, alpha=0.8, cmap='viridis', levels=20)
        ax1.scatter(X_plot[feature1], X_plot[feature2], c=y_plot, 
                cmap='viridis', edgecolors='black', s=30, alpha=0.7)
        ax1.set_xlabel(feature1, fontsize=12)
        ax1.set_ylabel(feature2, fontsize=12)
        ax1.set_title('决策树特征空间分层\n(颜色表示预测收盘价)', fontsize=14, fontweight='bold')
        plt.colorbar(contour, ax=ax1, label='预测收盘价')
        
        # 图2: 实际值 vs 预测值的散点图
        ax2 = axes[0, 1]
        y_pred_2d = dt_visual.predict(X_plot)
        scatter = ax2.scatter(y_plot, y_pred_2d, c=y_plot, 
                            cmap='coolwarm', alpha=0.7, s=50)
        ax2.plot([y_plot.min(), y_plot.max()], [y_plot.min(), y_plot.max()], 
                'r--', linewidth=2, label='完美预测线')
        ax2.set_xlabel('实际收盘价', fontsize=12)
        ax2.set_ylabel('预测收盘价', fontsize=12)
        ax2.set_title('实际值 vs 预测值', fontsize=14, fontweight='bold')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        plt.colorbar(scatter, ax=ax2, label='实际收盘价')
        
        # 图3: 残差分析
        ax3 = axes[1, 0]
        residuals = y_plot - y_pred_2d
        scatter_res = ax3.scatter(y_pred_2d, residuals, c=abs(residuals), 
                                cmap='Reds', alpha=0.7, s=50)
        ax3.axhline(y=0, color='red', linestyle='--', linewidth=2)
        ax3.set_xlabel('预测收盘价', fontsize=12)
        ax3.set_ylabel('残差', fontsize=12)
        ax3.set_title('预测残差分析', fontsize=14, fontweight='bold')
        ax3.grid(True, alpha=0.3)
        plt.colorbar(scatter_res, ax=ax3, label='残差绝对值')
        
        # 图4: 特征重要性分析
        ax4 = axes[1, 1]
        features = [feature1, feature2]
        importances = dt_visual.feature_importances_
        bars = ax4.bar(features, importances, color=['skyblue', 'lightcoral'], 
                    edgecolor='black', alpha=0.8)
        ax4.set_ylabel('特征重要性', fontsize=12)
        ax4.set_title('开盘价 vs 成交量 特征重要性', fontsize=14, fontweight='bold')
        ax4.grid(True, alpha=0.3, axis='y')
        
        # 在柱状图上添加数值
        for bar, importance in zip(bars, importances):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{importance:.3f}', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('./4-2.png')
        plt.show()
        
        # 打印统计信息
        print("\n" + "="*60)
        print("特征空间分层分析统计")
        print("="*60)
        print(f"使用的特征: {feature1}, {feature2}")
        print(f"特征重要性: {feature1}: {importances[0]:.4f}, {feature2}: {importances[1]:.4f}")
        print(f"R² Score: {r2_score(y_plot, y_pred_2d):.4f}")
        print(f"MSE: {mean_squared_error(y_plot, y_pred_2d):.4f}")
# 使用示例
def main():
    # 设置中文字体
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
    plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号
    # 假设你的数据格式如下：
    # data = pd.read_csv('你的纯碱期货数据.csv')
    
    # 创建示例数据（在实际使用时请替换为你的真实数据）
    np.random.seed(42)
    # n_samples = 1000
    # dates = pd.date_range('2020-01-01', periods=n_samples, freq='D')
    
    # # 生成更真实的期货价格数据（带有趋势和波动）
    # base_price = 2000
    # trend = np.cumsum(np.random.normal(0.5, 2, n_samples))  # 缓慢上涨趋势
    # noise = np.random.normal(0, 20, n_samples)
    # close_prices = base_price + trend + noise
    
    # example_data = pd.DataFrame({
    #     '日期': dates,
    #     '开盘价(元)': close_prices + np.random.normal(0, 10, n_samples),
    #     '最高价(元)': close_prices + np.random.normal(5, 8, n_samples),
    #     '最低价(元)': close_prices + np.random.normal(-5, 8, n_samples),
    #     '收盘价(元)': close_prices,
    #     '涨跌幅': np.random.normal(0, 0.02, n_samples),  # 日涨跌幅在±2%左右
    #     '成交额(百万)': np.random.uniform(50, 500, n_samples),
    #     '成交量': np.random.randint(10000, 100000, n_samples)
    # })
    
    # # 确保最高价 >= 收盘价 >= 最低价
    # example_data['最高价(元)'] = np.maximum(example_data['最高价(元)'], example_data['收盘价(元)'])
    # example_data['最低价(元)'] = np.minimum(example_data['最低价(元)'], example_data['收盘价(元)'])
    
    print("数据基本信息:")
    col=['日期', '开盘价(元)', '最高价(元)', '最低价(元)', '收盘价(元)', '涨跌幅', '成交额(百万)', '成交量']
    print(data[col].describe())
    
    # 初始化模型
    model = FuturesDecisionTree(data[col], target_col='收盘价(元)')
    
    # 准备数据
    model.prepare_data(lookback_days=10)
    
    # 训练模型
    model.train_model(use_grid_search=False, max_depth=8, min_samples_split=15, min_samples_leaf=8)
    
    # 评估模型
    metrics = model.evaluate_model()
    
    # 可视化结果
    model.plot_predictions()
    model.plot_feature_importance(top_n=15)
    model.plot_decision_tree(max_depth=3)
    model.plot_feature_space_stratification()
    
    return model, metrics

if __name__ == "__main__":
    model, metrics = main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

def plot_pruned_tree_performance( max_leaf_nodes_range=range(2, 51), cv_folds=5, figsize=(15, 10)):
    """绘制修剪后回归树表现：树大小 vs 均方误差"""
    
    # 准备数据
    feature1 = '开盘价(元)'
    feature2 = '成交量'
    target_col = '收盘价(元)'
    
    df = data.copy()
    X = df[[feature1, feature2]].dropna()
    y = df[target_col].loc[X.index]
    
    # 分割数据
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )
    
    # 存储结果
    train_errors = []
    cv_errors = []
    test_errors = []
    leaf_nodes = []
    
    print("训练不同大小的决策树...")
    
    for max_leaf_nodes in max_leaf_nodes_range:
        # 训练决策树
        dt = DecisionTreeRegressor(
            max_leaf_nodes=max_leaf_nodes,
            random_state=42
        )
        dt.fit(X_train, y_train)
        
        # 计算训练误差
        y_train_pred = dt.predict(X_train)
        train_error = mean_squared_error(y_train, y_train_pred)
        
        # 计算交叉验证误差
        cv_scores = cross_val_score(dt, X_train, y_train, 
                                   cv=cv_folds, scoring='neg_mean_squared_error')
        cv_error = -cv_scores.mean()
        
        # 计算测试误差
        y_test_pred = dt.predict(X_test)
        test_error = mean_squared_error(y_test, y_test_pred)
        
        # 存储结果
        train_errors.append(train_error)
        cv_errors.append(cv_error)
        test_errors.append(test_error)
        leaf_nodes.append(dt.get_n_leaves())
        
        if max_leaf_nodes % 10 == 0:
            print(f"完成 {max_leaf_nodes}/{max(max_leaf_nodes_range)}")
    
    # 找到最优的树大小
    best_cv_idx = np.argmin(cv_errors)
    best_leaf_nodes = leaf_nodes[best_cv_idx]
    best_cv_error = cv_errors[best_cv_idx]
    
    # 创建图形
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    makersize=8
    
    # 图1: 树大小 vs 均方误差
    ax1.plot(leaf_nodes, train_errors, 'b-', linewidth=2, label='训练集MSE', marker='o', markersize=makersize)
    ax1.plot(leaf_nodes, cv_errors, 'r-', linewidth=2, label='交叉验证MSE', marker='s', markersize=makersize)
    ax1.plot(leaf_nodes, test_errors, 'g-', linewidth=2, label='测试集MSE', marker='^', markersize=makersize)
    
    # 标记最优点
    ax1.axvline(x=best_leaf_nodes, color='red', linestyle='--', alpha=0.7, 
                label=f'最优树大小: {best_leaf_nodes}个叶节点')
    ax1.plot(best_leaf_nodes, best_cv_error, 'ro', markersize=makersize, 
             markeredgecolor='black', markeredgewidth=2)
    
    ax1.set_xlabel('树的大小 - 终端节点数目', fontsize=26)
    ax1.set_ylabel('均方误差 (MSE)', fontsize=26)
    ax1.set_title('修剪后回归树表现\n树大小 vs 均方误差', fontsize=26, fontweight='bold')
    # ax1.legend()
    ax1.legend(fontsize=16, loc='best')
    ax1.grid(True, alpha=0.3)
    
    # 图2: 偏差-方差权衡分析
    # 计算偏差和方差的近似值
    bias_squared = np.array(train_errors)  # 训练误差近似偏差平方
    variance = np.array(cv_errors) - np.array(train_errors)  # 交叉验证与训练误差差近似方差
    
    ax2.plot(leaf_nodes, bias_squared, 'k-', linewidth=2, label='偏差平方', marker='o', markersize=makersize)
    ax2.plot(leaf_nodes, variance, 'm-', linewidth=2, label='方差', marker='s', markersize=makersize)
    ax2.plot(leaf_nodes, cv_errors, 'r-', linewidth=2, label='总误差(交叉验证)', marker='^', markersize=makersize)
    
    ax2.axvline(x=best_leaf_nodes, color='red', linestyle='--', alpha=0.7)
    ax2.plot(best_leaf_nodes, cv_errors[best_cv_idx], 'ro', markersize=makersize, 
             markeredgecolor='black', markeredgewidth=2)
    
    ax2.set_xlabel('树的大小 - 终端节点数目', fontsize=26)
    ax2.set_ylabel('误差分量', fontsize=26)
    ax2.set_title('偏差-方差权衡分析', fontsize=26, fontweight='bold')
    # ax2.legend()
    ax2.legend(fontsize=16, loc='best')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # 保存图片
    filename = './4-4.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"图片已保存为: {filename}")
    
    plt.show()
    
    # 打印详细分析
    print("\n" + "="*70)
    print("修剪后回归树性能分析")
    print("="*70)
    print(f"最优树大小: {best_leaf_nodes} 个叶节点")
    print(f"最优交叉验证MSE: {best_cv_error:.4f}")
    print(f"对应的训练集MSE: {train_errors[best_cv_idx]:.4f}")
    print(f"对应的测试集MSE: {test_errors[best_cv_idx]:.4f}")
    
    # 分析过拟合和欠拟合区域
    small_tree_idx = len(leaf_nodes) // 4
    large_tree_idx = len(leaf_nodes) - 1
    
    print(f"\n区域分析:")
    print(f"小树 (叶节点数={leaf_nodes[small_tree_idx]}):")
    print(f"  训练MSE: {train_errors[small_tree_idx]:.4f}, 验证MSE: {cv_errors[small_tree_idx]:.4f}")
    print(f"  差距: {cv_errors[small_tree_idx] - train_errors[small_tree_idx]:.4f} (欠拟合)")
    
    print(f"大树 (叶节点数={leaf_nodes[large_tree_idx]}):")
    print(f"  训练MSE: {train_errors[large_tree_idx]:.4f}, 验证MSE: {cv_errors[large_tree_idx]:.4f}")
    print(f"  差距: {cv_errors[large_tree_idx] - train_errors[large_tree_idx]:.4f} (过拟合)")

def plot_detailed_pruning_analysis(max_depth_range=range(1, 11), figsize=(16, 12)):
    """详细的修剪分析：不同深度和叶节点限制"""
    
    feature1 = '开盘价(元)'
    feature2 = '成交量'
    target_col = '收盘价(元)'
    
    df =data.copy()
    X = df[[feature1, feature2]].dropna()
    y = df[target_col].loc[X.index]
    
    # 分割数据
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )
    
    # 存储结果
    results = []
    
    print("进行详细修剪分析...")
    
    for max_depth in max_depth_range:
        for max_leaf_nodes in [5, 10, 20, 30, 50, None]:
            dt = DecisionTreeRegressor(
                max_depth=max_depth,
                max_leaf_nodes=max_leaf_nodes,
                random_state=42
            )
            dt.fit(X_train, y_train)
            
            # 计算各种误差
            y_train_pred = dt.predict(X_train)
            train_error = mean_squared_error(y_train, y_train_pred)
            
            cv_scores = cross_val_score(dt, X_train, y_train, 
                                      cv=5, scoring='neg_mean_squared_error')
            cv_error = -cv_scores.mean()
            
            y_test_pred = dt.predict(X_test)
            test_error = mean_squared_error(y_test, y_test_pred)
            
            results.append({
                'max_depth': max_depth,
                'max_leaf_nodes': max_leaf_nodes if max_leaf_nodes else '无限制',
                'leaf_nodes': dt.get_n_leaves(),
                'train_error': train_error,
                'cv_error': cv_error,
                'test_error': test_error
            })
    
    # 转换为DataFrame
    results_df = pd.DataFrame(results)
    
    # 创建图形
    fig, axes = plt.subplots(2, 2, figsize=figsize)
    
    # 图1: 不同深度的表现
    for depth in max_depth_range:
        depth_data = results_df[results_df['max_depth'] == depth]
        axes[0, 0].plot(depth_data['leaf_nodes'], depth_data['train_error'], 
                       marker='o', label=f'深度{depth}-训练')
        axes[0, 0].plot(depth_data['leaf_nodes'], depth_data['cv_error'], 
                       marker='s', linestyle='--', label=f'深度{depth}-验证')
    
    axes[0, 0].set_xlabel('终端节点数目')
    axes[0, 0].set_ylabel('均方误差')
    axes[0, 0].set_title('不同深度下树大小 vs 误差')
    axes[0, 0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    axes[0, 0].grid(True, alpha=0.3)
    
    # 图2: 最优配置热力图
    pivot_table = results_df.pivot_table(values='cv_error', 
                                       index='max_depth', 
                                       columns='max_leaf_nodes', 
                                       aggfunc='mean')
    
    im = axes[0, 1].imshow(pivot_table.values, cmap='viridis', aspect='auto')
    axes[0, 1].set_xticks(range(len(pivot_table.columns)))
    axes[0, 1].set_xticklabels(pivot_table.columns)
    axes[0, 1].set_yticks(range(len(pivot_table.index)))
    axes[0, 1].set_yticklabels(pivot_table.index)
    axes[0, 1].set_xlabel('最大叶节点数')
    axes[0, 1].set_ylabel('最大深度')
    axes[0, 1].set_title('交叉验证误差热力图')
    plt.colorbar(im, ax=axes[0, 1], label='MSE')
    
    # 图3: 泛化差距分析
    results_df['generalization_gap'] = results_df['cv_error'] - results_df['train_error']
    
    scatter = axes[1, 0].scatter(results_df['leaf_nodes'], results_df['generalization_gap'],
                               c=results_df['max_depth'], cmap='viridis', s=50, alpha=0.7)
    axes[1, 0].axhline(y=0, color='red', linestyle='--', alpha=0.7)
    axes[1, 0].set_xlabel('终端节点数目')
    axes[1, 0].set_ylabel('泛化差距 (验证MSE - 训练MSE)')
    axes[1, 0].set_title('泛化差距分析')
    axes[1, 0].grid(True, alpha=0.3)
    plt.colorbar(scatter, ax=axes[1, 0], label='树深度')
    
    # 图4: 最优模型表现
    best_idx = results_df['cv_error'].idxmin()
    best_config = results_df.loc[best_idx]
    
    configs_to_show = ['小树(欠拟合)', '最优树', '大树(过拟合)']
    small_idx = results_df['leaf_nodes'].idxmin()
    large_idx = results_df['leaf_nodes'].idxmax()
    
    comparison_data = [
        results_df.loc[small_idx],
        best_config,
        results_df.loc[large_idx]
    ]
    
    x_pos = np.arange(len(configs_to_show))
    width = 0.25
    
    for i, metric in enumerate(['train_error', 'cv_error', 'test_error']):
        values = [data[metric] for data in comparison_data]
        axes[1, 1].bar(x_pos + i*width, values, width, 
                      label=['训练', '验证', '测试'][i])
    
    axes[1, 1].set_xlabel('模型配置')
    axes[1, 1].set_ylabel('均方误差')
    axes[1, 1].set_title('不同配置模型性能比较')
    axes[1, 1].set_xticks(x_pos + width)
    axes[1, 1].set_xticklabels(configs_to_show)
    axes[1, 1].legend()
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # 保存图片
    filename = 'detailed_pruning_analysis.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"详细分析图片已保存为: {filename}")
    
    plt.show()
    
    # 打印最优配置
    print("\n" + "="*70)
    print("最优修剪配置")
    print("="*70)
    print(f"最优配置: 深度={best_config['max_depth']}, 叶节点限制={best_config['max_leaf_nodes']}")
    print(f"实际叶节点数: {best_config['leaf_nodes']}")
    print(f"训练MSE: {best_config['train_error']:.4f}")
    print(f"交叉验证MSE: {best_config['cv_error']:.4f}")
    print(f"测试MSE: {best_config['test_error']:.4f}")
    
    return results_df

# 在类中添加这些方法
plot_pruned_tree_performance()
plot_detailed_pruning_analysis()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.inspection import permutation_importance
import warnings
warnings.filterwarnings('ignore')

class FuturesDecisionTree:
    """纯碱期货决策树回归模型"""
    
    def __init__(self, data, target_col='收盘价(元)', test_size=0.2, random_state=42):
        self.data = data.copy()
        self.target_col = target_col
        self.test_size = test_size
        self.random_state = random_state
        self.scaler = StandardScaler()
        self.model = None
        self.feature_importance = None
        
    def create_features(self, lookback_days=10):
        """创建技术指标特征"""
        print("创建特征工程...")
        
        df = self.data.copy()
        
        # 确保日期列正确处理
        if '日期' in df.columns:
            df['日期'] = pd.to_datetime(df['日期'])
            df = df.sort_values('日期').reset_index(drop=True)
        
        # 价格特征
        df['高低价比'] = df['最高价(元)'] / df['最低价(元)']
        df['开盘收盘价比'] = df['开盘价(元)'] / df['收盘价(元)']
        
        # 移动平均线特征
        for window in [5, 10, 20]:
            df[f'{window}日移动平均'] = df['收盘价(元)'].rolling(window=window).mean()
            df[f'价格移动平均比_{window}'] = df['收盘价(元)'] / df[f'{window}日移动平均']
            df[f'价格标准差_{window}'] = df['收盘价(元)'].rolling(window=window).std()
        
        # 动量指标
        df['5日动量'] = df['收盘价(元)'] - df['收盘价(元)'].shift(5)
        df['10日动量'] = df['收盘价(元)'] - df['收盘价(元)'].shift(10)
        df['5日价格变化率'] = (df['收盘价(元)'] - df['收盘价(元)'].shift(5)) / df['收盘价(元)'].shift(5) * 100
        df['10日价格变化率'] = (df['收盘价(元)'] - df['收盘价(元)'].shift(10)) / df['收盘价(元)'].shift(10) * 100
        
        # 波动率特征
        df['5日波动率'] = df['收盘价(元)'].rolling(window=5).std() / df['收盘价(元)'].rolling(window=5).mean()
        df['10日波动率'] = df['收盘价(元)'].rolling(window=10).std() / df['收盘价(元)'].rolling(window=10).mean()
        
        # 价格位置特征
        df['20日价格位置'] = (df['收盘价(元)'] - df['收盘价(元)'].rolling(20).min()) / \
                           (df['收盘价(元)'].rolling(20).max() - df['收盘价(元)'].rolling(20).min())
        
        # 涨跌幅特征
        if '涨跌幅' in df.columns:
            df['涨跌幅_绝对值'] = np.abs(df['涨跌幅'])
            df['涨跌幅_5日平均'] = df['涨跌幅'].rolling(5).mean()
            df['涨跌幅_10日平均'] = df['涨跌幅'].rolling(10).mean()
        
        # 滞后特征
        for lag in range(1, lookback_days + 1):
            df[f'收盘价滞后{lag}天'] = df['收盘价(元)'].shift(lag)
        
        # 成交量特征
        if '成交量' in df.columns:
            df['成交量_5日平均'] = df['成交量'].rolling(5).mean()
            df['成交量比率'] = df['成交量'] / df['成交量_5日平均']
            # 价格成交量相关性
            df['价量相关性_5日'] = df['收盘价(元)'].rolling(5).corr(df['成交量'])
        
        # 成交额特征
        if '成交额(百万)' in df.columns:
            df['成交额_5日平均'] = df['成交额(百万)'].rolling(5).mean()
            df['成交额比率'] = df['成交额(百万)'] / df['成交额_5日平均']
            df['平均成交价格'] = df['成交额(百万)'] * 1000000 / df['成交量']  # 计算平均成交价格
        
        # 目标变量：未来1天的收盘价
        df['目标收盘价'] = df['收盘价(元)'].shift(-1)
        
        # 删除含有NaN的行
        df = df.dropna()
        
        print(f"特征创建完成，最终数据形状: {df.shape}")
        return df
    
    def prepare_data(self, lookback_days=10):
        """准备训练数据"""
        print("准备训练数据...")
        
        # 创建特征
        df = self.create_features(lookback_days)
        
        # 分离特征和目标变量
        exclude_cols = ['目标收盘价', '日期', self.target_col]
        feature_cols = [col for col in df.columns if col not in exclude_cols]
        self.feature_cols = feature_cols
        
        X = df[feature_cols]
        y = df['目标收盘价']
        
        print(f"特征数量: {len(feature_cols)}")
        print("前10个特征:", feature_cols[:10])
        
        # 数据分割（时间序列分割，不随机打乱）
        split_idx = int(len(X) * (1 - self.test_size))
        
        self.X_train = X.iloc[:split_idx]
        self.X_test = X.iloc[split_idx:]
        self.y_train = y.iloc[:split_idx]
        self.y_test = y.iloc[split_idx:]
        
        # # 特征标准化
        # self.X_train_scaled = self.scaler.fit_transform(self.X_train)
        # self.X_test_scaled = self.scaler.transform(self.X_test)
        self.X_train_scaled = self.X_train
        self.X_test_scaled = self.X_test
        print(f"训练集大小: {self.X_train_scaled.shape}")
        print(f"测试集大小: {self.X_test_scaled.shape}")
        print(f"训练集时间范围: {df['日期'].iloc[0]} 到 {df['日期'].iloc[split_idx-1]}")
        print(f"测试集时间范围: {df['日期'].iloc[split_idx]} 到 {df['日期'].iloc[-1]}")
        
        return self.X_train_scaled, self.X_test_scaled, self.y_train, self.y_test
    
    def tune_hyperparameters(self):
        """超参数调优"""
        print("进行超参数调优...")
        
        param_grid = {
            'max_depth': [3, 5, 7, 10, 15, None],
            'min_samples_split': [2, 5, 10, 20],
            'min_samples_leaf': [1, 2, 5, 10],
            'max_features': ['auto', 'sqrt', 'log2', None]
        }
        
        dt = DecisionTreeRegressor(random_state=self.random_state)
        grid_search = GridSearchCV(
            dt, param_grid, cv=5, scoring='neg_mean_squared_error', 
            n_jobs=-1, verbose=1
        )
        
        grid_search.fit(self.X_train_scaled, self.y_train)
        
        print(f"最佳参数: {grid_search.best_params_}")
        print(f"最佳交叉验证分数: {-grid_search.best_score_:.4f}")
        
        self.best_params = grid_search.best_params_
        return grid_search.best_params_
    
    def train_model(self, use_grid_search=True, **kwargs):
        """训练决策树模型"""
        print("训练决策树回归模型...")
        
        if use_grid_search and hasattr(self, 'best_params'):
            model_params = self.best_params
        else:
            # 默认参数
            model_params = {
                'max_depth': 10,
                'min_samples_split': 10,
                'min_samples_leaf': 5,
                'max_features': 'sqrt',
                'random_state': self.random_state
            }
            model_params.update(kwargs)
        
        self.model = DecisionTreeRegressor(**model_params)
        self.model.fit(self.X_train_scaled, self.y_train)
        
        # 预测
        self.y_train_pred = self.model.predict(self.X_train_scaled)
        self.y_test_pred = self.model.predict(self.X_test_scaled)
        
        # 计算特征重要性
        self.feature_importance = pd.DataFrame({
            'feature': self.feature_cols,
            'importance': self.model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        return self.model
    
    def evaluate_model(self):
        """评估模型性能"""
        print("\n模型性能评估:")
        print("=" * 50)
        
        # 计算各项指标
        train_mse = mean_squared_error(self.y_train, self.y_train_pred)
        test_mse = mean_squared_error(self.y_test, self.y_test_pred)
        train_mae = mean_absolute_error(self.y_train, self.y_train_pred)
        test_mae = mean_absolute_error(self.y_test, self.y_test_pred)
        train_r2 = r2_score(self.y_train, self.y_train_pred)
        test_r2 = r2_score(self.y_test, self.y_test_pred)
        
        # 计算方向准确性（涨跌方向预测）
        train_direction = np.mean((np.diff(self.y_train.values) * np.diff(self.y_train_pred)) > 0)
        test_direction = np.mean((np.diff(self.y_test.values) * np.diff(self.y_test_pred)) > 0)
        
        # 计算平均预测误差百分比
        train_mape = np.mean(np.abs((self.y_train - self.y_train_pred) / self.y_train)) * 100
        test_mape = np.mean(np.abs((self.y_test - self.y_test_pred) / self.y_test)) * 100
        
        print(f"训练集 - MSE: {train_mse:.4f}, MAE: {train_mae:.4f}, R²: {train_r2:.4f}")
        print(f"训练集 - MAPE: {train_mape:.2f}%, 方向准确率: {train_direction:.4f}")
        print(f"测试集 - MSE: {test_mse:.4f}, MAE: {test_mae:.4f}, R²: {test_r2:.4f}")
        print(f"测试集 - MAPE: {test_mape:.2f}%, 方向准确率: {test_direction:.4f}")
        
        return {
            'train_mse': train_mse,
            'test_mse': test_mse,
            'train_mae': train_mae,
            'test_mae': test_mae,
            'train_r2': train_r2,
            'test_r2': test_r2,
            'train_mape': train_mape,
            'test_mape': test_mape,
            'train_direction': train_direction,
            'test_direction': test_direction
        }
    
    def plot_predictions(self):
        """绘制预测结果"""
        plt.figure(figsize=(15, 12))
        
        # 训练集预测结果
        plt.subplot(3, 1, 1)
        plt.plot(self.y_train.values, label='实际收盘价', alpha=0.8, linewidth=1)
        plt.plot(self.y_train_pred, label='预测收盘价', alpha=0.8, linewidth=1)
        plt.title('训练集: 实际值 vs 预测值')
        plt.xlabel('样本索引')
        plt.ylabel('价格(元)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # 测试集预测结果
        plt.subplot(3, 1, 2)
        plt.plot(self.y_test.values, label='实际收盘价', alpha=0.8, linewidth=1)
        plt.plot(self.y_test_pred, label='预测收盘价', alpha=0.8, linewidth=1)
        plt.title('测试集: 实际值 vs 预测值')
        plt.xlabel('样本索引')
        plt.ylabel('价格(元)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # 散点图
        plt.subplot(3, 2, 5)
        plt.scatter(self.y_test, self.y_test_pred, alpha=0.6, s=20)
        plt.plot([self.y_test.min(), self.y_test.max()], 
                [self.y_test.min(), self.y_test.max()], 'r--', lw=2)
        plt.xlabel('实际收盘价(元)')
        plt.ylabel('预测收盘价(元)')
        plt.title('测试集: 实际值 vs 预测值散点图')
        plt.grid(True, alpha=0.3)
        
        # 残差图
        plt.subplot(3, 2, 6)
        residuals = self.y_test - self.y_test_pred
        plt.scatter(self.y_test_pred, residuals, alpha=0.6, s=20)
        plt.axhline(y=0, color='r', linestyle='--')
        plt.xlabel('预测收盘价(元)')
        plt.ylabel('残差')
        plt.title('测试集残差图')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def plot_feature_importance(self, top_n=15):
        """绘制特征重要性图"""
        if self.feature_importance is None:
            print("请先训练模型!")
            return
        
        plt.figure(figsize=(12, 10))
        
        top_features = self.feature_importance.head(top_n)
        
        plt.subplot(2, 1, 1)
        plt.barh(range(len(top_features)), top_features['importance'], color='skyblue')
        plt.yticks(range(len(top_features)), top_features['feature'], fontsize=16, fontweight='bold')
        plt.xlabel('特征重要性', fontsize=26, fontweight='bold')
        plt.title(f'纯碱期货决策树特征重要性 (Top {top_n})', fontsize=26, fontweight='bold')
        plt.legend(fontsize=16, loc='best')
        plt.gca().invert_yaxis()
        plt.grid(True, alpha=0.3, axis='x')
        
        plt.subplot(2, 1, 2)
        # 累积重要性
        cumulative_importance = top_features['importance'].cumsum()
        plt.plot(range(1, len(top_features) + 1), cumulative_importance, 'o-', color='orange')
        plt.axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='80% 重要性')
        plt.axhline(y=0.9, color='green', linestyle='--', alpha=0.7, label='90% 重要性')
        plt.xlabel('特征数量', fontsize=26, fontweight='bold')
        plt.ylabel('累积重要性', fontsize=26, fontweight='bold')
        plt.title('特征累积重要性', fontsize=26, fontweight='bold')

    # ax1.legend()
        plt.legend(fontsize=16, loc='best')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('./4-5.png')
        plt.show()
        
        print(f"\n前{top_n}个重要特征:")
        print(top_features.round(4))
        
        # 计算达到80%和90%重要性所需的特征数量
        cumulative_all = self.feature_importance['importance'].cumsum()
        n_80 = len(cumulative_all[cumulative_all <= 0.8]) + 1
        n_90 = len(cumulative_all[cumulative_all <= 0.9]) + 1
        print(f"达到80%重要性所需特征数: {n_80}")
        print(f"达到90%重要性所需特征数: {n_90}")
    
    def plot_decision_tree(self, max_depth=3):
        """绘制决策树结构（限制深度以便可视化）"""
        if self.model is None:
            print("请先训练模型!")
            return
        
        plt.figure(figsize=(20, 10))
        plot_tree(self.model, 
                 feature_names=self.feature_cols,
                 filled=True, 
                 rounded=True,
                 max_depth=max_depth,
                 fontsize=26,
                 proportion=True)
        # plt.title(f'纯碱期货决策树结构 (最大深度: {max_depth})')
        # plt.title(f'纯碱期货价格预测决策树 (深度限制: {max_depth})', 
        #                 fontsize=26, fontweight='bold', pad=30)
        plt.tight_layout()
        
        plt.show()


    def plot_feature_space_stratification(self, feature1='开盘价(元)', feature2='成交量', 
                                        target_col='收盘价(元)', figsize=(16, 12)):
        """绘制基于特征空间分层的预测图（开盘价 vs 成交量）"""
        if self.model is None:
            print("请先训练模型!")
            return
        
        # 准备数据
        df = self.data.copy()
        
        # 选择两个特征和目标变量
        X_plot = df[[feature1, feature2]].dropna()
        y_plot = df[target_col].loc[X_plot.index]
        
        # 训练一个专门用于可视化的决策树
        dt_visual = DecisionTreeRegressor(max_depth=4, random_state=42)
        dt_visual.fit(X_plot, y_plot)
        
        # 创建网格数据
        x_min, x_max = X_plot[feature1].min() - 1, X_plot[feature1].max() + 1
        y_min, y_max = X_plot[feature2].min() - 1, X_plot[feature2].max() + 1
        
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                            np.linspace(y_min, y_max, 200))
        
        # 预测整个网格
        Z = dt_visual.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        # 创建图形
        fig, axes = plt.subplots(2, 2, figsize=figsize)
        
        # 图1: 决策树分区边界
        ax1 = axes[0, 0]
        contour = ax1.contourf(xx, yy, Z, alpha=0.8, cmap='viridis', levels=20)
        ax1.scatter(X_plot[feature1], X_plot[feature2], c=y_plot, 
                cmap='viridis', edgecolors='black', s=30, alpha=0.7)
        ax1.set_xlabel(feature1, fontsize=12)
        ax1.set_ylabel(feature2, fontsize=12)
        ax1.set_title('决策树特征空间分层\n(颜色表示预测收盘价)', fontsize=14, fontweight='bold')
        plt.colorbar(contour, ax=ax1, label='预测收盘价')
        
        # 图2: 实际值 vs 预测值的散点图
        ax2 = axes[0, 1]
        y_pred_2d = dt_visual.predict(X_plot)
        scatter = ax2.scatter(y_plot, y_pred_2d, c=y_plot, 
                            cmap='coolwarm', alpha=0.7, s=50)
        ax2.plot([y_plot.min(), y_plot.max()], [y_plot.min(), y_plot.max()], 
                'r--', linewidth=2, label='完美预测线')
        ax2.set_xlabel('实际收盘价', fontsize=12)
        ax2.set_ylabel('预测收盘价', fontsize=12)
        ax2.set_title('实际值 vs 预测值', fontsize=14, fontweight='bold')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        plt.colorbar(scatter, ax=ax2, label='实际收盘价')
        
        # 图3: 残差分析
        ax3 = axes[1, 0]
        residuals = y_plot - y_pred_2d
        scatter_res = ax3.scatter(y_pred_2d, residuals, c=abs(residuals), 
                                cmap='Reds', alpha=0.7, s=50)
        ax3.axhline(y=0, color='red', linestyle='--', linewidth=2)
        ax3.set_xlabel('预测收盘价', fontsize=12)
        ax3.set_ylabel('残差', fontsize=12)
        ax3.set_title('预测残差分析', fontsize=14, fontweight='bold')
        ax3.grid(True, alpha=0.3)
        plt.colorbar(scatter_res, ax=ax3, label='残差绝对值')
        
        # 图4: 特征重要性分析
        ax4 = axes[1, 1]
        features = [feature1, feature2]
        importances = dt_visual.feature_importances_
        bars = ax4.bar(features, importances, color=['skyblue', 'lightcoral'], 
                    edgecolor='black', alpha=0.8)
        ax4.set_ylabel('特征重要性', fontsize=12)
        ax4.set_title('开盘价 vs 成交量 特征重要性', fontsize=14, fontweight='bold')
        ax4.grid(True, alpha=0.3, axis='y')
        
        # 在柱状图上添加数值
        for bar, importance in zip(bars, importances):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{importance:.3f}', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('./4-2.png')
        plt.show()
        
        # 打印统计信息
        print("\n" + "="*60)
        print("特征空间分层分析统计")
        print("="*60)
        print(f"使用的特征: {feature1}, {feature2}")
        print(f"特征重要性: {feature1}: {importances[0]:.4f}, {feature2}: {importances[1]:.4f}")
        print(f"R² Score: {r2_score(y_plot, y_pred_2d):.4f}")
        print(f"MSE: {mean_squared_error(y_plot, y_pred_2d):.4f}")
# 使用示例
def main():
    # 设置中文字体
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
    plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号
    # 假设你的数据格式如下：
    # data = pd.read_csv('你的纯碱期货数据.csv')
    
    # 创建示例数据（在实际使用时请替换为你的真实数据）
    np.random.seed(42)
    # n_samples = 1000
    # dates = pd.date_range('2020-01-01', periods=n_samples, freq='D')
    
    # # 生成更真实的期货价格数据（带有趋势和波动）
    # base_price = 2000
    # trend = np.cumsum(np.random.normal(0.5, 2, n_samples))  # 缓慢上涨趋势
    # noise = np.random.normal(0, 20, n_samples)
    # close_prices = base_price + trend + noise
    
    # example_data = pd.DataFrame({
    #     '日期': dates,
    #     '开盘价(元)': close_prices + np.random.normal(0, 10, n_samples),
    #     '最高价(元)': close_prices + np.random.normal(5, 8, n_samples),
    #     '最低价(元)': close_prices + np.random.normal(-5, 8, n_samples),
    #     '收盘价(元)': close_prices,
    #     '涨跌幅': np.random.normal(0, 0.02, n_samples),  # 日涨跌幅在±2%左右
    #     '成交额(百万)': np.random.uniform(50, 500, n_samples),
    #     '成交量': np.random.randint(10000, 100000, n_samples)
    # })
    
    # # 确保最高价 >= 收盘价 >= 最低价
    # example_data['最高价(元)'] = np.maximum(example_data['最高价(元)'], example_data['收盘价(元)'])
    # example_data['最低价(元)'] = np.minimum(example_data['最低价(元)'], example_data['收盘价(元)'])
    
    print("数据基本信息:")
    col1=data.describe().columns

    print(data[col1].describe())
    
    # 初始化模型
    model = FuturesDecisionTree(data[col1], target_col='收盘价(元)')
    
    # 准备数据
    model.prepare_data(lookback_days=10)
    
    # 训练模型
    model.train_model(use_grid_search=False, max_depth=8, min_samples_split=15, min_samples_leaf=8)
    
    # 评估模型
    metrics = model.evaluate_model()
    
    # 可视化结果
    model.plot_predictions()
    model.plot_feature_importance(top_n=15)
    model.plot_decision_tree(max_depth=3)
    model.plot_feature_space_stratification()
    
    return model, metrics

if __name__ == "__main__":
    model, metrics = main()
```
