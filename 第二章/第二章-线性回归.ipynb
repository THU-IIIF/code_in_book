### 2.1 线性回归

```python
import statsmodels.formula.api as smf

import matplotlib.pyplot as plt
import  pandas as pd
data=pd.read_csv('../data/book_ipo.csv',index_col=0)### 读取数据
data
```

```python
var='revision'
model = smf.ols('ir ~'+var, data = data)
result = model.fit()

plt.figure(figsize = (10, 8), dpi = 80)
plt.scatter(data['revision'],result.fittedvalues)
plt.scatter(data['revision'], data['ir'])#画拟合线

plt.savefig('./output/ipo_'+var+'.png',dpi=1000,bbox_inches = 'tight')
plt.show()
result.summary()
```

```python

var='nasdaq15'
model = smf.ols('ir ~'+var, data = data)
result = model.fit()

plt.figure(figsize = (10, 8), dpi = 80)
plt.scatter(data['revision'],result.fittedvalues)
plt.scatter(data['revision'], data['ir'])#画拟合线

plt.show()
result.summary()
```

```python
var='lrassets'
model = smf.ols('ir ~'+var, data = data)
result = model.fit()

plt.figure(figsize = (10, 8), dpi = 80)
plt.scatter(data[var],result.fittedvalues)
plt.scatter(data[var], data['ir'])#画拟合线

plt.show()
result.summary()
```

```python
var='vc'
model = smf.ols('ir ~'+var, data = data)
result = model.fit()

plt.figure(figsize = (10, 8), dpi = 80)
plt.scatter(data[var],result.fittedvalues)
plt.scatter(data[var], data['ir'])#画拟合线

plt.show()
result.summary()
```

```python
model = smf.ols('ir ~ nasdaq15_pos + lrassets + vc', data = data)
result = model.fit()

plt.figure(figsize = (10, 8), dpi = 80)
plt.scatter(result.fittedvalues, result.resid)

plt.show()
```
