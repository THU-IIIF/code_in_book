```python
rolling=5
start=origin['weekly'].min()
test0=pd.DataFrame([])
# for k in range(start+rolling,x):
for k in range(rolling+start,origin['weekly'].max()+1):
    print(k)
    train=origin[(origin['weekly']>=k-rolling)&(origin['weekly']<k)]
    train=train.dropna()
    test=origin[origin['weekly']==k]
#     test=test.dropna()
# train
    x=varX
    X_train=train[x]
    y_train=train[y]
    X = sm.add_constant(X_train)
    model = sm.OLS(y_train, X).fit()
#     test['pred']=model.predict( sm.add_constant(test[x]))
    if 'const' in model.params:
        test['pred']=model.params['const'] 
    else:
        test['pred']=0
    for coef in range(len(x)):
#         print(coef)
        test['pred']=test['pred']+test[x[coef]]*model.params[coef]
    test0=pd.concat([test0,test])
test
```

```python
test0=test0[test0['weekly']>=103]
test0=test0[test0['weekly']<=245]
test0=test0.reset_index(drop=True)
test0['pos']=1*(test0['pred']>0)-1*(test0['pred']<0)
test0['profit']=(test0['pos']==1)*test0['re']-(test0['pos']==-1)*test0['re']
test0['asset']=test0['profit'].cumsum()
# test0['asset'].plot()
# test0['weekly']
print('OLS模型最终累计收益率',round(test0['asset'].iloc[-1],3))
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error,r2_score
 
def evaluation(y_test, y_predict,n=363,p=100):
    
    mae = mean_absolute_error(y_test, y_predict)
    mse = mean_squared_error(y_test, y_predict)
    rmse = np.sqrt(mean_squared_error(y_test, y_predict))
    mape=(abs(y_predict -y_test)/ abs(y_test)).mean()
    r_2=r2_score(y_test, y_predict)
    
    r2_adj=1-((1-r2_score(y_test,y_predict))*(n-1))/(n-p-1)
    corr=np.corrcoef(y_test, y_predict)[0,1]
    return mae, rmse, mape,corr,r_2,r2_adj #mse
df_eval=pd.DataFrame(columns=['MAE','RMSE','MAPE','corr','R2','R2_adj'])
```
